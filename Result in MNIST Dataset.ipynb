# Result in MNIST Dataset

### 1. Import the model and data

```python
import dbm2
import rbm
import matplotlib.pyplot as plt
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

### 2. Set the Parameters

```python
EPOCH = 30
MINI_BATCH = 100
LR=1e-3
H = (500, 1000)
```

### 3. RBM

```python
rbm_agent = rbm.RBM(name='demo', n_v=mnist.train.images.shape[1], n_h=H[0], epoch=EPOCH, learning_rate=LR, mini_batch=MINI_BATCH, save_path='rbm.ckpt')

rbm_agent.fit(mnist.train.images)

rbm_agent.save()

original = mnist.test.next_batch(MINI_BATCH)[0]
reconstructed, _ = rbm_agent.transform(original)
```

Visualize the sample generated from RBM. The images shown are probability of reconstructed visible units given the original visible unit. The second graph shows the original data.

```python
original = mnist.test.next_batch(MINI_BATCH)[0]
reconstructed, hidden = rbm_agent.transform(original)

nfigs = 5
fig1 = plt.figure(figsize=(10,10))
for i in range(nfigs ** 2):
     ax = fig1.add_subplot(nfigs, nfigs, i+1)
     ax.imshow(reconstructed[i].reshape(28, 28))

fig2 = plt.figure(figsize=(10,10))
for i in range(nfigs ** 2):
     ax = fig2.add_subplot(nfigs, nfigs, i+1)
     ax.imshow(hidden[i].reshape(28, 28))
```

### DBM(2 layers of hidden units)

```python
dbm = dbm2.DBM(n_v=784, n_hs=H, mini_batch=100, learning_rate=1e-3, epoch=20, gibbs_step=5, mean_field_step=5)

dbm.fit(mnist.train.images)

dbm.save()
```

Visualize the sample generated from DBM. The images shown are probability of reconstructed visible units given the original visible unit. The second graph shows the original data.

```python
original = mnist.test.next_batch(MINI_BATCH)[0]

reconstructed, hidden1, hidden2 = dbm.transform(original)
nfigs = 5
fig3 = plt.figure(figsize=(10,10))
for i in range(nfigs ** 2):
     ax = fig3.add_subplot(nfigs, nfigs, i+1)
     ax.imshow(reconstructed[i].reshape(28, 28))
    
nfigs = 5
fig3 = plt.figure(figsize=(10,10))
for i in range(nfigs ** 2):
     ax = fig3.add_subplot(nfigs, nfigs, i+1)
     ax.imshow(original[i].reshape(28, 28))
```
